```python
import os
```

```python
index_stock_path = r"directory/index_stock"
derivative = r"directory/derivative"
```

```python
os.makedirs(index_stock_path, exist_ok=True)
os.makedirs(derivative, exist_ok=True)
```

```python
for year in range(2014,2025):
    for month in range(1,13):
        for day in range(1,32):
            print(f"For year: {year} / month: {month} / day {day}")
            try:
                # Download bhavcopy
                bhavcopy_save(date(year,month,day), index_stock_path)

                # Download bhavcopy for futures and options
                bhavcopy_fo_save(date(year,month,day), derivative)
            except Exception as e:
                print(year, month, day, e)
```
