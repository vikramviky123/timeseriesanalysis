```python
import os
import json
from datetime import datetime


import yfinance as yf

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
nifty50_list = []

with open('nifty50.txt','r') as f:
    for stk in f.readlines():
        nifty50_list.append(stk.split('\n')[0])
        # print(stk)
```

```python
nifty100_list = []

with open('nifty100.txt','r') as f:
    for stk in f.readlines():
        nifty100_list.append(stk.split('\n')[0])
        # print(stk)
```

```python
nifty50_df = pd.DataFrame()
for each_stk in nifty50_list:
    temp_df = yf.download(str(each_stk), start='2014-01-01', end='2024-05-15')
    temp_df['stk_type'] = str(each_stk)
    
    nifty50_df = pd.concat([nifty50_df,temp_df], axis=0)
```

```python
nifty100_df = pd.DataFrame()
for each_stk in nifty100_list:
    temp_df = yf.download(str(each_stk), start='2014-01-01', end='2024-05-15')
    temp_df['stk_type'] = str(each_stk)
    
    nifty100_df = pd.concat([nifty100_df,temp_df], axis=0)
```

```python
nifty50_df = nifty50_df.reset_index()
nifty100_df = nifty100_df.reset_index()
```

```python
nifty50_df['Year'] = nifty50_df['Date'].dt.year
nifty50_df['Month'] = nifty50_df['Date'].dt.month
nifty50_df['Day'] = nifty50_df['Date'].dt.day

nifty100_df['Year'] = nifty100_df['Date'].dt.year
nifty100_df['Month'] = nifty100_df['Date'].dt.month
nifty100_df['Day'] = nifty100_df['Date'].dt.day
```

```python
nifty50_df.to_csv("nifty50.csv", index=False)
nifty100_df.to_csv("nifty100.csv", index=False)
```

```python

# end date data is excluded 
df_nifty = yf.download('^NSEI', start='2014-01-01', end='2024-05-15')
df_bnf = yf.download('^NSEBANK', start='2014-01-01', end='2024-05-15')

df_nifty = df_nifty.reset_index()
df_bnf = df_bnf.reset_index()
```

```python
import pandas as pd
```

```python
merge_df = pd.merge(df_nifty,df_bnf, how='left', on='Date')
```

```python
merge_df.columns
```

```python
merge_df[['Close_x','Close_y']].corr()
```

```python
merge_df['nifty_hl_pc'] = ((merge_df['High_x'] - merge_df['Low_x'])/merge_df['Low_x'])*100
merge_df['nifty_oc_pc'] = ((merge_df['Close_x'] - merge_df['Open_x'])/merge_df['Open_x'])*100
merge_df['bnf_hl_pc'] = ((merge_df['High_y'] - merge_df['Low_y'])/merge_df['Low_y'])*100
merge_df['bnf_oc_pc'] = ((merge_df['Close_y'] - merge_df['Open_y'])/merge_df['Open_x'])*100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(25,6))
sns.boxplot(merge_df['bnf_oc_pc'], orient='h')
# plt.xticks([i * 0.5 for i in range(int(21 / 0.4))])
plt.show()
```

```python
plt.figure(figsize=(25,6))
sns.boxplot(merge_df['nifty_oc_pc'], orient='h')
# plt.xticks([i * 0.3 for i in range(int(10 / 0.3))])
plt.show()
```

```python
merge_df['Year'] = merge_df['Date'].dt.year
merge_df['Month'] = merge_df['Date'].dt.month
merge_df['Day'] = merge_df['Date'].dt.day
```

```python
merge_df['Year'].unique()
```

```python
for each_year in merge_df['Year'].unique():
    new_df = merge_df[merge_df['Year'] == each_year]
    
    plt.figure(figsize=(25, 6))
    sns.boxplot(x='bnf_oc_pc', data=new_df)

    # Adjusting x-axis ticks
    # plt.xticks([i * 0.4 for i in range(int(22 / 0.4))])
    plt.title(str(each_year))

    plt.show()
```

```python
g = sns.FacetGrid(merge_df, col="Year")
g.map(sns.boxplot, "bnf_pc");
```

```python
import numpy as np
year_df_nifty = merge_df.groupby(['Year','Month'])['nifty_oc_pc'].agg([np.min,
                                            lambda x: np.percentile(x, 25),
                                            np.mean,
                                            np.median,                                          
                                             lambda x: np.percentile(x, 75),
                                              lambda x: np.percentile(x, 90),
                                              lambda x: np.percentile(x, 95),
                                              lambda x: np.percentile(x, 99),
                                              np.max,
                                              ]).reset_index()
year_df_nifty['type'] = 'nifty'
year_df_bnf = merge_df.groupby(['Year','Month'])['bnf_oc_pc'].agg([np.min,
                                            lambda x: np.percentile(x, 25),
                                            np.mean,
                                            np.median,                                          
                                             lambda x: np.percentile(x, 75),
                                              lambda x: np.percentile(x, 90),
                                              lambda x: np.percentile(x, 95),
                                              lambda x: np.percentile(x, 99),
                                              np.max,
                                              ]).reset_index()
year_df_bnf['type'] = 'bnf'
```

```python
import pandas as pd
```

```python
comb_df = pd.concat([year_df_nifty,year_df_bnf], axis= 0).reset_index(drop=True)
```

```python
comb_df
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(x = range(1,126),y=comb_df[comb_df['type'] =='nifty']['mean'], label='nifty')
sns.lineplot(x = range(1,126),y=comb_df[comb_df['type'] =='bnf']['mean'], label='bnf')
plt.xticks(range(1,126))
plt.show()
```

```python
for each_year in merge_df['Year'].unique():
    new_df = comb_df[comb_df['Year'] == each_year]
    plt.figure(figsize=(15,4))
    sns.lineplot(x=new_df['Month'],y=new_df['mean'], hue = new_df['type'])
    plt.title(str(each_year))
    plt.xticks(range(1,13))
    plt.show()
```
